```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
_0_05 = r'C:\Users\Anak\Documents\Downloads\Experiment2_0.05.csv'

x = pd.read_csv(_0_05)
x.index = ['BM','SHeG', 'SHeP' , 'SHeGaP', 'SHeGoP', 'SHeGnP', 'SHePnG']
x.plot.bar()
```

```python
bm_GD = [x.loc['BM']['GD']]
bm_PGD = [x.loc['BM']['PGD']]
bm_PGDP = [x.loc['BM']['PGDP']]
```

```python
x_np = x.to_numpy()
x_dict = []
graph_type = ['GD', 'PGD', 'PGDP']
qualified_nodes = ['BM','SHeG', 'SHeP', 'SHeGoP','SHeGaP' , 'SHeGnP', 'SHePnG']
for i, q in zip(x_np, qualified_nodes):
    
    for j,g in zip(i, graph_type):
        tmp = {}
        tmp['auc'] =j
        tmp['graph_type'] = g
        tmp['qualified_nodes'] = q
        x_dict.append(tmp)
        print(tmp)
# print(x_dict)
x_df = pd.DataFrame(x_dict)
```

```python
ax = sns.barplot(x="qualified_nodes", y="auc", hue="graph_type", data=x_df)
plt.plot(bm_GD * 7, linewidth=2, color='blue')
plt.plot(bm_PGD * 7, linewidth=2, color='orange')
plt.plot(bm_PGDP * 7, linewidth=2, color='green')
ax.set_ylim(0.6, 0.95)
plt.title('threshold=0.05')
```
