## read emb file (generated by node2vec) to pandas DataFrame (take a look at the emb_File; first columns is not a part of dataframe)

### First Solution

```python
s = """1,2
a,b,d,e
a,b,d,e
a,b,d,e
a,b,d,e
"""
df = pd.read_csv(StringIO(s), skiprows=0,header=1)
new_df = pd.DataFrame(np.insert(df.values, 0,values=list(df.columns), axis=0))
new_df.set_index(0, inplace=True)
del new_df.index.name

print(new_df)
```
